```python

import pandas as pd
import numpy as np
from google.colab import files

print("Upload your CSV file")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("Dataset Loaded Successfully!")
df.head(1300)
```

```python
# Show missing values count
print("Missing Values:\n", df.isnull().sum())

# Fill numeric → mean
# Fill categorical → mode
df_clean = df.copy()

for col in df_clean.columns:
    if df_clean[col].dtype != 'object':     # numeric
        df_clean[col].fillna(df_clean[col].mean(), inplace=True)
    else:                                   # categorical
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

print("\nMissing Values Handled Successfully!")


```

```python

def remove_duplicates(df):
    before = df.shape[0]
    df_cleaned = df.drop_duplicates()
    after = df_cleaned.shape[0]
    print(f"Duplicates removed: {before - after}")
    return df_cleaned

df = remove_duplicates(df)

from scipy import stats

df_zscore = df.select_dtypes(include=[np.number])    # numeric columns only
z_scores = np.abs(stats.zscore(df_zscore, nan_policy='omit'))

# threshold = 3 (common for z-score)
outliers = np.where(z_scores > 3)

print("Z-Score Outliers Found:", len(outliers[0]))

# Remove outliers
df_no_outliers_z = df[(z_scores < 3).all(axis=1)]

print("Shape After Removing Z-Score Outliers:", df_no_outliers_z.shape)
```

```python
print("Final Cleaned Data Shape:", df.shape)
df.head(1300)
```

```python
df_clean.to_csv("cleaned_dataset.csv", index=False)
files.download("cleaned_dataset.csv")

print("Cleaned CSV Saved & Downloaded Successfully!")
```
